### Integral equations

The definite integral is given by 
1. Inline $\int_{a}^{b} f(x) \,dx$, with larger text $ \displaystyle\int_a^b f(x) \,dx $
2. For an integral displayed on its own line
$$
\int_{a}^{b} f(x) \,dx = F(b) - F(a)
$$
3. Multiple integrals
$$ 
\iint_V \mu(u,v) \,du\,dv 
$$

###  Differential Equations

1. For a differential term within a line of text: Eg. The derivative of $y$ with respect to $x$ is $\frac{dy}{dx}$.
2. For a differential equation displayed on its own line,     
$$
\frac{d^2y}{dx^2} + k^2y = 0
$$
3. Partial derivative $ \displaystyle\frac{\partial f}{\partial x} $

$$ \frac{\partial u}{\partial t} = \alpha \left( \frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2} \right) $$

### Integral of a differential equation:
$$
\int \left( \frac{dy}{dx} \right)^2 \,dx = C
$$

The solution to the ordinary differential equation $$ \frac{\mathrm{d}y}{\mathrm{d}x} + 2xy = x $$ is found by integrating a factor. The resulting particular solution might be given by the integral: $$ y = e^{-x^2} \int_0^x e^{t^2}t\,dt $$
